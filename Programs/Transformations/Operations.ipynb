```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Generate some example data
np.random.seed(0)
data_left_skewed = -np.random.exponential(scale=2, size=1000)

# Function to apply log transformation
def log_transform(data):
    return np.log(data + 1)

# Function to plot data and its log transformation
def plot_data_and_transformation(data, title):
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    
    # Plot original data
    sns.histplot(data, kde=True, ax=axes[0])
    axes[0].set_title(f'Original Data - {title}')
    axes[0].set_xlabel('Value')
    axes[0].set_ylabel('Frequency')
    axes[0].text(0.95, 0.95, f'Skewness: {skew(data):.2f}', 
                 verticalalignment='top', horizontalalignment='right', 
                 transform=axes[0].transAxes, color='black', fontsize=12)
    
    # Plot log-transformed data
    transformed_data = log_transform(data)
    sns.histplot(transformed_data, kde=True, ax=axes[1])
    axes[1].set_title(f'Log-Transformed Data - {title}')
    axes[1].set_xlabel('Value')
    axes[1].set_ylabel('Frequency')
    axes[1].text(0.95, 0.95, f'Skewness: {skew(transformed_data):.2f}', 
                 verticalalignment='top', horizontalalignment='right', 
                 transform=axes[1].transAxes, color='black', fontsize=12)
    
    plt.tight_layout()
    plt.show()

# Check skewness and plot data
print(f"Skewness of left-skewed data: {skew(data_left_skewed)}")
plot_data_and_transformation(data_left_skewed, 'Left-Skewed')

transformed_data_left_skewed = log_transform(data_left_skewed)
print(f"Skewness of log-transformed left-skewed data: {skew(transformed_data_left_skewed)}")
```
